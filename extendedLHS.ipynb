```python
import numpy as np
import matplotlib.pyplot as plt
from libs.latinhypercube import LHS_extendable
```

```python
n_samples = 5

X1range = (-5., 10.) # must be float
X2range = (0., 15.)
space = [X1range, X2range]

lhs = LHS_extendable(criterion="maximin", iterations=10000)
x = lhs.generate(space, n_samples) # generate sample

x
```

```python
n_samples = 10

y = lhs.generate(space, n_samples, existing_samples=x)
y
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(np.array(y)[:,0], np.array(y)[:,1], 'o', c='cornflowerblue', label = 'extended')
ax.plot(np.array(x)[:,0], np.array(x)[:,1], 'o', c='indianred', label = 'original')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.legend()
```
